```python
# headers: headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'}
# url: http://www.pythonhow.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/

import requests
from bs4 import BeautifulSoup

r = requests.get("http://www.pythonhow.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/", 
                headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
c = r.content

soup = BeautifulSoup(c,"html.parser")
#print(soup.prettify())

all = soup.find_all("div",{"class":"propertyRow"})

# list does not have a find all method, but you can go to each item in the list to apply it in this case
# find_all returns a list
all[0].find("h4",{"class":"propPrice"}).text.replace("\n","").replace(" ","")
```

```python
all
```

```python
# extract the price for each house on display
l = []
for item in all:
    d = {}

    # address is span to two tags
    d["Address"] = item.find_all("span",{"class":"propAddressCollapse"})[0].text
    d["Locality"] = item.find_all("span",{"class":"propAddressCollapse"})[1].text
    # price
    d["Price"] = item.find("h4",{"class":"propPrice"}).text.replace("\n","").replace(" ","")
    
    # beds
    try:
        d["Beds"] = item.find("span",{"class":"infoBed"}).find("b").text
    except:
        d["Beds"] = None
    
    # square feet
    try:
        d["Area"] = item.find("span",{"class":"infoSqFt"}).find("b").text
    except:
        d["Area"] = None
    
    # full baths
    try:
        d["Full Baths"] = item.find("span",{"class":"infoValueFullBath"}).find("b").text
    except:
        d["Full Baths"] = None
    
    # half baths
    try:
        d["Half Baths"] = item.find("span",{"class":"infoValueHalfBath"}).find("b").text
    except:
        d["Half Baths"] = None
        
    # tricky to get the property features, need to see how everything is organized
    for column_group in item.find_all("div",{"class":"columnGroup"}):
        #print(column_group)
        for featureGroup, featureName in zip(column_group.find_all("span",{"class":"featureGroup"}),column_group.find_all("span",{"class":"featureName"})):       
            d[featureGroup.text.replace(u'\xa0', u' ').split(':',1)[0]] = featureName.text.split(',',1)[0]
            
    l.append(d)
    # print(" ")
```

```python
# to do it for all the pages, get the page number count
page_num = soup.find_all("a",{"class":"Page"})[-1].text
print(page_num)
```

```python
# to cycle through different pages, the pages for this url are done as such
base_url = "http://pyclass.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/t=0&s="
real_estate_list = []
for page in range(0,int(page_num)*10,10):
    print(base_url+str(page)+".html")
    r = requests.get(base_url+str(page)+".html", headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
    c = r.content
    soup = BeautifulSoup(c,"html.parser")
    soup.prettify()
    all = soup.find_all("div",{"class":"propertyRow"})
    for item in all:
        d = {}

        # address is span to two tags
        d["Address"] = item.find_all("span",{"class":"propAddressCollapse"})[0].text
        
        try:
            d["Locality"] = item.find_all("span",{"class":"propAddressCollapse"})[1].text
        except:
            d["Locality"] = None
        
        # price
        d["Price"] = item.find("h4",{"class":"propPrice"}).text.replace("\n","").replace(" ","")

        # beds
        try:
            d["Beds"] = item.find("span",{"class":"infoBed"}).find("b").text
        except:
            d["Beds"] = None

        # square feet
        try:
            d["Area"] = item.find("span",{"class":"infoSqFt"}).find("b").text
        except:
            d["Area"] = None

        # full baths
        try:
            d["Full Baths"] = item.find("span",{"class":"infoValueFullBath"}).find("b").text
        except:
            d["Full Baths"] = None

        # half baths
        try:
            d["Half Baths"] = item.find("span",{"class":"infoValueHalfBath"}).find("b").text
        except:
            d["Half Baths"] = None

        # tricky to get the property features, need to see how everything is organized
        for column_group in item.find_all("div",{"class":"columnGroup"}):
            #print(column_group)
            for featureGroup, featureName in zip(column_group.find_all("span",{"class":"featureGroup"}),column_group.find_all("span",{"class":"featureName"})):       
                d[featureGroup.text.replace(u'\xa0', u' ').split(':',1)[0]] = featureName.text.split(',',1)[0]

        real_estate_list.append(d)
    
```

```python
real_estate_df = pandas.DataFrame(real_estate_list)
```
